```python
from graphviz import Digraph

# Create a directed graph
dot = Digraph(comment="High-Level Booking Pipeline", format="png")
dot.attr(rankdir='TB', size='8')

# Nodes
dot.node('A', 'Customer Interaction Layer\n(Web Form / Chat)\nAWS S3 + JS Frontend',
         shape='box', style='filled', fillcolor='#ffebcd')
dot.node('B', 'LLM Processing\n(OpenAI GPT-4o-mini)\nExtract structured booking data',
         shape='box', style='filled', fillcolor='#e6e6fa')
dot.node('C', 'Job Management Backend\n(FastAPI)',
         shape='box', style='filled', fillcolor='#f0fff0')
dot.node('D', 'Database\n(AWS DynamoDB)',
         shape='cylinder', style='filled', fillcolor='#f5f5f5')
dot.node('E', 'Technician Assignment Logic\n(Skill, Location, Workload, Time Availability)',
         shape='box', style='filled', fillcolor='#ffe4e1')
dot.node('F', 'Booking Confirmation\n+ Technician Details\n(Return to Frontend)',
         shape='box', style='filled', fillcolor='#fafad2')

# Edges
dot.edge('A', 'B', label='Booking request (NL text)')
dot.edge('B', 'C', label='Structured booking data (JSON)')
dot.edge('C', 'D', label='Store job record')
dot.edge('C', 'E', label='Trigger assignment logic')
dot.edge('E', 'D', label='Update job with assigned technician')
dot.edge('E', 'F', label='Return assigned tech details')
dot.edge('F', 'A', label='Display confirmation to customer')

# Display in notebook
dot
```
